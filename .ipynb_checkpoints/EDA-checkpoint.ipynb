```python
import numpy as np
import pandas as pd
```

```python
import nltk
from nltk.corpus import PlaintextCorpusReader, stopwords
from nltk import FreqDist
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures, TrigramCollocationFinder, TrigramAssocMeasures
```

```python
import os
path = 'data/Wheel of Time/txt'
file_list = os.listdir(path)
```

```python
df = pd.read_csv('data/details.csv')
df
```

```python
def read_file(filename):
    corpus = PlaintextCorpusReader(path, filename)
    print(filename, " completed")
    return corpus
```

```python
df['corpus'] = df['filename'].apply(read_file)
```

```python
df['total_chapters'] = df['num_chapters'] + df['prologue'] + df['epilogue']
```

```python
df
```

```python
df['word_count'] = df['corpus'].apply(lambda x: len(x.words()))
df['vocab_size'] = df['corpus'].apply(lambda x: len(set(x.words())))
```

```python
df['num_of_sents'] = df['corpus'].apply(lambda x: len(x.sents()))
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df['words'] = df['corpus'].apply(lambda x: x.words())
df['words_lean'] = df['words'].apply(lambda x: [word for word in x if word.lower() not in stop_words and word.isalnum()])
```

```python
df['freq_dist'] = df['words_lean'].apply(lambda x: FreqDist(x))
```

```python
df['freq_dist'].apply(lambda x: x.most_common(10))
```

```python
for fdist in list(df['freq_dist']):
    fdist.plot(15, cumulative=False)
```

```python
df['long_words'] = df['words_lean'].apply(lambda x: [word for word in x if len(word) > 14])
```

```python
df['frequent_bigrams'] = df['words_lean'].apply(lambda x: BigramCollocationFinder.from_words(x).nbest(BigramAssocMeasures.likelihood_ratio, 25))
```

```python
df['words_lean'].apply(lambda x: TrigramCollocationFinder.from_words(x).nbest(TrigramAssocMeasures.likelihood_ratio, 25))[12]
```
